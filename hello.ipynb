```python
def hi():
  print("hi")
  import matplotlib.pyplot as plt
  import numpy as np

  #nofeatures, m, alpha, ceiling, k = 100, 3.6, 0.11, 98, 23
  nofeatures, m, alpha, ceiling, k = 200, 3.5, 0.11, 98, 23

  p = np.arange(0, nofeatures, 1)

  y = np.piecewise(
      p,
      [p <= k, p > k],
      [
          lambda x: m * x,
          lambda x: ceiling - (ceiling - m*k)*np.exp(-alpha*(x - k))
      ]
  )


  fig = plt.figure(figsize=(10, 5))
  plt.hlines(ceiling, 0, nofeatures, "r", "dashed", label = "Max Accuracy Ceiling (a_max)")
  plt.vlines(k, 0, 100, "b", "dashed", label = "k Features")
  plt.xticks(visible=False)
  plt.yticks(visible=False)
  plt.ylabel("Accuracy (a_p)")
  plt.xlabel("Number of Features (p)")
  plt.plot(p, y, label = "Accuracy for each p feature", color = "k")
  plt.legend()
  plt.show()
hi()
```
